```python
#Import Necessary Library's
import pandas as pd
import json
from pathlib import Path
```

```python
#Import Data
data_folder = Path("./Data/")
FreedomFile = data_folder / "economic_freedom_index2019_data.csv"
economicFreedomDF = pd.read_csv(FreedomFile,encoding='latin-1')

economicFreedomDF.set_index("Country Name", inplace = True)
economicFreedomDF2 = economicFreedomDF.drop(["CountryID","WEBNAME"], axis = 1)
economicFreedomDF2 = economicFreedomDF2.dropna(subset=["World Rank"])
economicFreedomDF2
```

```python
economicFreedomDF3 = economicFreedomDF2.drop(economicFreedomDF2.columns.difference(["2019 Score", "GDP (Billions, PPP)", "GDP per Capita (PPP)","Financial Freedom", "Population (Millions)", "Unemployment (%)", "Inflation (%)"]), 1)
economicFreedomDF3
```
